```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?id=13840;type=tournament"
```

website response

```python
res=requests.get(url)
```

```python
res
```

parsing

```python
soup=BeautifulSoup(res.text)

```

```python
soup
```

```python
players=soup.select(".data-link")
```

```python
players
```

```python
player_names=[]
```

```python

for item in players:
    player_names.append(item.text)
    link=".data-link"
```

```python
player_names
```

```python
score=soup.select("td+ td b")
```

```python
score
```

```python
high_score=[]
```

```python

for item in score:
    high_score.append(item.text)
```

```python
high_score
```

```python
no_match=[]
```

```python

match=soup.select(".data2 .left+ .padAst")
```

```python
match
```

```python
for item in match:
    no_match.append(item.text)
```

```python
no_match
```

```python
highest_sixes=[]
```

```python

sixes=soup.select("td:nth-child(14)")
```

```python
sixes
```

```python
for item in sixes:
    highest_sixes.append(item.text)
```

```python
highest_sixes
```
